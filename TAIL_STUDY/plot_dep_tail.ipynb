```python
#### Plot DEP tail events

import sys
import pandas as pd
sys.path.append('../')
from functions import plot_hits, plot_deco_reco
import matplotlib.pyplot as plt
```

```python
# load data
data_DECO = pd.read_hdf('../data/tail_evts.h5', key = 'DECO')
display(data_DECO)
```

```python
for evt, df in data_DECO.groupby('event'):
    print('=' * 20)
    print(f'Event: {evt}')
    print(f'Energy: {df.E.sum():.2f} MeV')
    plot_hits(df)
    plt.show()
```
